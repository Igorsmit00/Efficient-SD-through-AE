## The Benefits of Auto-Encoding in Subgroup Discovery

Start by importing packages and functions from local files

```python
# Package imports
import datetime
import matplotlib.pyplot as plt

# Local imports
from dataImporter import getData
from utilityFunctions import standardize
from autoEncoder import autoEncode
from beamSearch import EMM, as_string
from adjPysubgroup import adjustedBestFirstSearch, adjustedDFS, adjustedApriori
from qualityMeasures import calc_result_bs, calc_result_ps
```

The next step is to import a dataset. The following options are available when using the local getData() function:

"Ionosphere", "Adult", "Soybean", "Arrhythmia", "Indoor" and "Mushroom"

  In this experiment, we will proceed with the "Adult" dataset:

```python
df, cat, num, features = getData("Ionosphere")
df_to_encode = df
```

For datasets where numerical attributes could have disproportionate influence on the autoencoder loss function, the following standardization function can be applied on the numerical columns:

```python
# df_to_encode = standardize(df, num)
```

The last step before auto-encoding can commence is to give a lower and upper estimate for the number of features

```python
l_est, r_est = 2, 12
```

We then proceed to auto-encode the selected DataFrame to determine the number of features the dataset should be reduced to by plotting the resulting MSE

```python
t1 = datetime.datetime.now()
mseList = []
for i in range(l_est, r_est+1) :
    df_en, mse = autoEncode(df_to_encode, cat, num, nFeatures = i, nEpochs=100)
    mseList.append(mse)
t2 = datetime.datetime.now()
plt.plot([i for i in range(l_est, r_est+1)], mseList)
plt.show()
running_time = t2-t1
running_time
```

From the figure, we conclude that the optimal number of features is 5. We thus proceed to produce a new dataframe with 5 attributes.

```python
df_en, _ = autoEncode(df_to_encode, cat, num, nFeatures = 5, nEpochs=100, verbose=False)
```

We can now run the subgroup discovery algorithms on the dataset. We start with the beam search algorithm:

```python
t1 = datetime.datetime.now()
output = EMM(100, 3, 100, [], df, features, 'target', ensure_diversity=True)
t2 = datetime.datetime.now()
running_time = t2-t1
running_time
```

```python
t1 = datetime.datetime.now()
features_en = [i for i in df_en.columns if 'cat' in i]
output_en = EMM(100, 3, 100, [], df_en, features_en, 'target', ensure_diversity=True)
t2 = datetime.datetime.now()
running_time = t2-t1
running_time
```

We can then summarize the results as follows:

```python
subgroups_n = [i for i in output.get_values()]
subgroups_en = [i for i in output_en.get_values()]
calc_result_bs(df_en, df, subgroups_en, subgroups_n)
```

Next, we proceed to use the BestFirstSearch algorithm:

```python
t1 = datetime.datetime.now()
results_df_n = adjustedBestFirstSearch(df)
t2 = datetime.datetime.now()
running_time = t2-t1
running_time
```

```python
t1 = datetime.datetime.now()
results_df_en = adjustedBestFirstSearch(df_en)
t2 = datetime.datetime.now()
running_time = t2-t1
running_time
```

Finally, we summarize the results for the BestFirstSearch algorithm:

```python
calc_result_ps(df_en, df, results_df_en, results_df_n)
```
